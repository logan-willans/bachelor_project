```python
cities = list(hometown['hometown_city'].unique())
cities
```

```python
city_dict = dict()

counter = 0

for city in cities:
    if city in city_dict:
        continue
    city_dict[city] = counter
    counter += 1

numbers = [city_dict[city] for city in cities]
```

```python
country_dict = {}

counter = 1

for country in countries:
    if country in country_dict:
        continue 
    country_dict[country] = counter
    counter += 1
    
```

```python
bachelorette.replace({'country': country_dict}, inplace=True)
```
